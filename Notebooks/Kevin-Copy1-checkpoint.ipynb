```python
def reglin(x,y):
    x = sm.add_constant(x.to_numpy())
    model = sm.OLS(y.to_numpy(),x).fit()
    B0 = model.params[0]
    B1 = model.params[1]
    x = x[:,1]
    x2 = np.linspace(x.min(),x.max(),100) 
    y_hat = x2 * B1 +B0
    #plt.scatter(x,y,alpha=1)
    plt.plot(x2,y_hat,'--g')   
    return [model, B0, B1]
```

```python
def get_keys(data, t_patern='s2_p%d_x1', y_patern='s2_p%d_y1'):
    t = []
    x = []
    keys = data.keys()
    for k in range(100):
        t0 = t_patern % (k+1)
        x0 = y_patern % (k+1)
        if not x0 in keys:
            x0 = 's1_p%d_y4' % (k+1)
        if not t0 in keys:
            t0 = 's1_p%d_x4' % (k+1)
        if (t0 in keys) & (x0 in keys):
            t.append(t0)
            x.append(x0)
        else:
            break
    return t, x
```
